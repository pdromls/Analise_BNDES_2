---

#### **1. Sumário**

Este relatório apresenta uma análise exploratória detalhada da base de dados de "operações de exportação pré-embarque" do BNDES. As principais descobertas indicam uma forte concentração dos recursos em múltiplos eixos: o setor da **Indústria** é o principal beneficiário, absorvendo um volume massivo de crédito em comparação com Comércio/Serviços e Agropecuária. Geograficamente, a maior parte do valor é destinada a empresas nas regiões **Sudeste e Sul**, com o estado de São Paulo liderando de forma isolada. A análise temporal revelou uma grande volatilidade no volume de crédito concedido anualmente, com um pico histórico no início da década de 2010. O estudo também identificou que os **grandes bancos comerciais** são os principais intermediários das operações e que empresas de **grande porte** concentram a maior parte do valor, embora exista uma dinâmica particular onde a mediana do valor para microempresas supera a de pequenas empresas.

---

#### **2. Introdução**

**2.1. Contexto:** Este estudo foi realizado com o objetivo de estudar, aplicar e praticar técnicas de análise de dados em uma base pública e real. A análise se concentra em um conjunto de dados do BNDES sobre operações de crédito, buscando compreender padrões de financiamento, identificar os principais setores e agentes beneficiados, e extrair tendências temporais.

**2.2. Objetivos:**
* Analisar a distribuição do crédito entre os diferentes setores econômicos (Indústria, Comércio/Serviços, Agropecuária).
* Mapear a distribuição geográfica dos recursos por estado (UF).
* Identificar tendências e picos no volume de crédito concedido ao longo dos anos.
* Compreender como o valor das operações se distribui entre empresas de diferentes portes (Micro, Pequeno, Médio e Grande).
* Determinar as principais instituições financeiras que atuam como intermediárias no repasse do crédito.

---

#### **3. Metodologia**

A análise foi conduzida utilizando técnicas de exploração e visualização de dados em Python, com o uso das bibliotecas Pandas, Matplotlib e Seaborn. Os resultados foram consolidados em dois notebooks Jupyter (`01_BNDES_DA.ipynb` e `02_BNDES_EDA.ipynb`), onde os dados foram tratados, explorados, agrupados, agregados e plotados em gráficos para identificar padrões, distribuições e tendências.

---

#### **4. Resultados e Análise**

**4.1. Distribuição do Crédito por Porte do Cliente**

A análise revela uma relação complexa entre o porte da empresa e o valor das operações. Embora a tendência geral mostre que empresas maiores acessam valores mais elevados, a mediana das operações para empresas **MICRO** é ligeiramente superior às operações das empresas da categoria **PEQUENA**. A variabilidade e a presença de operações de altíssimo valor (outliers) aumentam drasticamente para as empresas de **GRANDE** porte.



**4.2. Análise por Setor Econômico**

Os dados indicam uma concentração predominante dos recursos na **INDÚSTRIA**. Com um volume total superior a R$125 bilhões, este setor se posiciona como o principal beneficiário. O setor de **COMÉRCIO/SERVIÇOS**, em segundo lugar, acumulou aproximadamente R$4.2 bilhões, enquanto a **AGROPECUÁRIA** figura na terceira posição.

Os dados indicam uma concentração predominante dos recursos na **INDÚSTRIA**. Com um volume total superior a R$125 bilhões, este setor se posiciona como o principal beneficiário, recebendo uma quantia ordens de magnitude maior que os demais. O setor de **COMÉRCIO/SERVIÇOS**, em segundo lugar, acumulou aproximadamente R$ 4.2 bilhões, enquanto a **AGROPECUÁRIA** figura na terceira posição.



**4.3. Evolução do Crédito ao Longo do Tempo**

A evolução do volume de crédito ao longo dos anos exibe um comportamento volátil. O período de maior atividade ocorreu no início da década de 2010, com um pico histórico em **2010**. Após esse auge, seguiu-se um período de instabilidade e tendência de queda, atingindo os níveis mais baixos por volta de 2017, com uma leve recuperação nos anos mais recentes.



**4.4. Distribuição Geográfica do Crédito**

A distribuição geográfica do crédito é marcadamente concentrada nas regiões Sudeste e Sul. O estado de **São Paulo (SP)** se destaca como o líder absoluto, com mais de R$ 72 bilhões em operações. Na sequência, destacam-se **Rio Grande do Sul (RS)**, **Santa Catarina (SC)** e **Minas Gerais (MG)**.



**4.5. Análise por Instituição Financeira Intermediadora**

O repasse do crédito é dominado pelos grandes bancos comerciais do país. **Banco Bradesco**, **Itaú Unibanco** e **Banco do Brasil** formam o grupo líder, intermediando, cada um, cerca de R$ 25 bilhões. É notável a presença de uma categoria **"SEM NOME"** entre as cinco primeiras, indicando que uma parcela relevante das operações é registrada na base de dados sem a identificação explícita do agente.




---

#### **5. Conclusões e Recomendações**

A análise exploratória proporcionou insights valiosos sobre a dinâmica do crédito de exportação pré-embarque do BNDES. Ficou evidente um perfil de alta concentração, tanto setorial (Indústria) quanto geográfica (Sul-Sudeste), e uma forte volatilidade ao longo do tempo. O programa atende majoritariamente grandes empresas industriais por meio dos principais bancos comerciais do país.

**Recomendações:**

**Para o BNDES:** Avaliar políticas de incentivo para diversificar o apoio a outros setores, como Comércio/Serviços e Agropecuária, e ampliar o alcance geográfico para estados fora do eixo Sul-Sudeste, a fim de democratizar o acesso ao crédito.

**Para Empresas Exportadoras:** Empresas de grande porte no setor industrial podem continuar a ver esta linha como uma fonte robusta de financiamento, focando o relacionamento com os bancos líderes (Bradesco, Itaú, BB). Empresas de outras regiões ou setores podem precisar buscar ativamente informações sobre como acessar esses recursos.

**Para Análises Futuras:** Investigar os fatores macroeconômicos (câmbio, PIB, balança comercial) que podem explicar a volatilidade anual no volume de crédito. Além disso, seria interessante analisar os dados da categoria "SEM NOME" para entender melhor a cadeia de repasse.

Os dados indicam uma concentração predominante dos recursos na **INDÚSTRIA**. Com um volume total superior a R$ 125 bilhões, este setor se posiciona como o principal beneficiário, recebendo uma quantia ordens de magnitude maior que os demais. O setor de **COMÉRCIO/SERVIÇOS**, em segundo lugar, acumulou aproximadamente R$ 4.2 bilhões, enquanto a **AGROPECUÁRIA** figura na terceira posição. Isso evidencia que, para esta linha de crédito específica, o foco do BNDES é majoritariamente industrial.
