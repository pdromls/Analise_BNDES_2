#### **1. Sumrio**

Este relatrio apresenta uma anlise exploratria detalhada da base de dados de "operaes de exportao pr-embarque" do BNDES. As principais descobertas indicam uma forte concentrao dos recursos em mltiplos eixos: o setor da **Indstria**  o principal beneficirio, absorvendo um volume massivo de crdito em comparao com Comrcio/Servios e Agropecuria. O estudo tambm identificou que os **grandes bancos comerciais** so os principais intermedirios das operaes e que empresas de **grande porte** concentram a maior parte do valor, embora exista uma dinmica particular onde a mediana do valor para microempresas supera a de pequenas empresas.

#### **4. Resultados e Anlise**

**4.1. Distribuio do Crdito por Porte do Cliente**

A anlise revela uma relao complexa entre o porte da empresa e o valor das operaes. Embora a tendncia geral mostre que empresas maiores acessam valores mais elevados, a mediana das operaes para empresas **MICRO**  ligeiramente superior s operaes das empresas da categoria **PEQUENA**. A variabilidade e a presena de operaes de altssimo valor (outliers) aumentam drasticamente para as empresas de **GRANDE** porte.



**4.2. Anlise por Setor Econmico**

Os dados indicam uma concentrao predominante dos recursos na **INDSTRIA**. Com um volume total superior a R$125 bilhes, este setor se posiciona como o principal beneficirio. O setor de **COMRCIO/SERVIOS**, em segundo lugar, acumulou aproximadamente R$4.2 bilhes, enquanto a **AGROPECURIA** figura na terceira posio.

**4.3. Evoluo do Crdito ao Longo do Tempo**

A evoluo do volume de crdito ao longo dos anos exibe um comportamento voltil. O perodo de maior atividade ocorreu no incio da dcada de 2010, com um pico histrico em **2010**. Aps esse auge, seguiu-se um perodo de instabilidade e tendncia de queda, atingindo os nveis mais baixos por volta de 2017, com uma leve recuperao nos anos mais recentes.



**4.4. Distribuio Geogrfica do Crdito**

A distribuio geogrfica do crdito  marcadamente concentrada nas regies Sudeste e Sul. O estado de **So Paulo (SP)** se destaca como o lder absoluto, com mais de R$ 72 bilhes em operaes. Na sequncia, destacam-se **Rio Grande do Sul (RS)**, **Santa Catarina (SC)** e **Minas Gerais (MG)**.



**4.5. Anlise por Instituio Financeira Intermediadora**

O repasse do crdito  dominado pelos grandes bancos comerciais do pas. **Banco Bradesco**, **Ita Unibanco** e **Banco do Brasil** formam o grupo lder, intermediando, cada um, cerca de R$ 25 bilhes.  notvel a presena de uma categoria **"SEM NOME"** entre as cinco primeiras, indicando que uma parcela relevante das operaes  registrada na base de dados sem a identificao explcita do agente.

A anlise exploratria proporcionou insights valiosos sobre a dinmica do crdito de exportao pr-embarque do BNDES.
